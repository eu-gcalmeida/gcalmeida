## Investiga√ß√£o Preliminar dos Eventos

### Eventos √∫nicos nas amostras

```python
# Dicion√°rio de nomes (PDG ID -> Nome)
pdg_map = {
    1: "d", 2: "u", 3: "s", 4: "c", 5: "b", 6: "t", -1: "anti-d", -2: "anti-u", -3: "anti-s", -4: "anti-c", -5: "anti-b", -6: "anti-t",
    11: "e-", 12: "nu_e", 13: "mu-", 14: "nu_mu", 15: "tau-", 16: "nu_tau", -11: "e+", -12: "anti_nu_e", -13: "mu+", -14: "anti_nu_mu", -15: "tau+", -16: "anti_nu_tau",
    21: "gluon", 22: "gamma", 23: "Z0", 24: "W+", -24: "W-", 25: "Higgs"
}

def pegar_nome(pid):
    return pdg_map.get(pid, str(pid))

def mostrar_eventos_unicos(lista_eventos, nome_arquivo):
    print(f"\n{'='*20} Analisando Eventos √önicos: {nome_arquivo} {'='*20}")

    # Conjunto para guardar as "assinaturas" que j√° vimos
    assinaturas_vistas = set()
    contador_tipos = 0

    for i, evento in enumerate(lista_eventos):
        iniciais = []
        intermediarias = []
        finais = []

        # 1. Classifica as part√≠culas deste evento espec√≠fico
        for p in evento.particles:
            nome_formatado = f"{pegar_nome(p.id)} ({p.id})"

            if p.status == -1:
                iniciais.append(nome_formatado)
            elif p.status == 2:
                intermediarias.append(nome_formatado)
            elif p.status == 1:
                finais.append(nome_formatado)

        # 2. Cria uma "assinatura" (tupla) para conferir se j√° vimos este tipo exato antes
        # Ordenamos as listas para que (u, anti-u) seja igual a (anti-u, u)
        assinatura = (tuple(sorted(iniciais)), tuple(sorted(intermediarias)), tuple(sorted(finais)))

        # 3. Se for uma novidade, imprime a tabela!
        if assinatura not in assinaturas_vistas:
            assinaturas_vistas.add(assinatura)
            contador_tipos += 1

            print(f"\nüîé Tipo de Evento #{contador_tipos} (Encontrado no √≠ndice {i})")

            # Cria o DataFrame apenas para este evento
            dict_dados = {
                "Part√≠culas Iniciais (-1)": pd.Series(iniciais),
                "Intermedi√°rias (2)": pd.Series(intermediarias),
                "Finais Detect√°veis (1)": pd.Series(finais)
            }

            df = pd.DataFrame(dict_dados)
            display(df.fillna(" ")) # Limpa os valores vazios (NaN)

    print(f"\n‚úÖ Total de varia√ß√µes √∫nicas encontradas: {contador_tipos}")

# --- Execu√ß√£o ---

if 'eventos_sinal' in globals():
    mostrar_eventos_unicos(eventos_sinal, "SINAL")
else:
    print("‚ö†Ô∏è Carregue 'eventos_sinal' primeiro.")

if 'eventos_fundo' in globals():
    mostrar_eventos_unicos(eventos_fundo, "FUNDO")
else:
    print("‚ö†Ô∏è Carregue 'eventos_fundo' primeiro.")
```

### Compilado das part√≠culas envolvidas nos eventos e seus status

```python
# Dicion√°rio simples para traduzir os IDs mais comuns (PDG ID -> Nome)
pdg_map = {
    1: "d", 2: "u", 3: "s", 4: "c", 5: "b", 6: "t", -1: "anti-d", -2: "anti-u", -3: "anti-s", -4: "anti-c", -5: "anti-b", -6: "anti-t",
    11: "e-", 12: "nu_e", 13: "mu-", 14: "nu_mu", 15: "tau-", 16: "nu_tau", -11: "e+", -12: "anti_nu_e", -13: "mu+", -14: "anti_nu_mu", -15: "tau+", -16: "anti_nu_tau",
    21: "gluon", 22: "gamma", 23: "Z0", 24: "W+", -24: "W-", 25: "Higgs"
}

def pegar_nome(pid):
    # Tenta pegar o nome no dicion√°rio, se n√£o achar, devolve o n√∫mero como string
    return pdg_map.get(pid, str(pid))

def gerar_tabela_resumo(lista_eventos, nome_processo):
    # Sets (conjuntos) para guardar apenas part√≠culas √∫nicas e evitar repeti√ß√£o
    iniciais = set()
    intermediarias = set()
    finais = set()

    # Varre todos os 1000 eventos
    for evento in lista_eventos:
        for p in evento.particles:
            nome_formatado = f"{pegar_nome(p.id)} ({p.id})"

            if p.status == -1:
                iniciais.add(nome_formatado)
            elif p.status == 2: # Em LHE, 2 geralmente √© resson√¢ncia/intermedi√°rio
                intermediarias.add(nome_formatado)
            elif p.status == 1:
                finais.add(nome_formatado)

    # Cria o DataFrame para exibi√ß√£o (lidando com tamanhos diferentes de listas)
    # Convertemos os sets para listas ordenadas
    dict_dados = {
        "Part√≠culas Iniciais (Status -1)": sorted(list(iniciais)),
        "Intermedi√°rias (Status 2)": sorted(list(intermediarias)),
        "Finais Detect√°veis (Status 1)": sorted(list(finais))
    }

    # Truque do Pandas para criar tabela com colunas de tamanhos diferentes
    df = pd.DataFrame.from_dict(dict_dados, orient='index').transpose()

    print(f"\n=== Tabela de Part√≠culas: {nome_processo} ===")
    display(df.fillna("-")) # Preenche vazios com tra√ßo

# --- Executando para Sinal e Fundo ---
# Certifique-se que voc√™ j√° leu os arquivos antes e guardou em listas!
# Exemplo: eventos_sinal = list(pylhe.read_lhe_with_attributes("sinal.lhe"))

if 'eventos_sinal' in globals():
    gerar_tabela_resumo(eventos_sinal, "Processo de sinal")
else:
    print("‚ö†Ô∏è Vari√°vel 'eventos_sinal' n√£o encontrada. Leia o arquivo primeiro.")

if 'eventos_fundo' in globals():
    gerar_tabela_resumo(eventos_fundo, "Processo de fundo")
else:
    print("‚ö†Ô∏è Vari√°vel 'eventos_fundo' n√£o encontrada. Leia o arquivo primeiro.")
```

## Histogramas das vari√°veis cinem√°ticas para as part√≠culas de estado final (exceto neutrinos)

```python
# Definir uma fun√ß√£o para extrair as v√°riaveis cin√©ticas e criar listas com os valores

def var_cinematicas (evento):
  pt_final = []       # lista do momento transversal
  eta_final = []      # lista da pseudorapidez
  phi_final = []      # lista do √¢ngulo azimutal
  for e in evento:
    for p in e.particles:
        # Objetivo 3: Apenas estado final (status = 1) e ignorar neutrinos
        # Dica: Neutrinos t√™m IDs 12, 14, 16 (e os antineutrinos negativos)
        is_neutrino = abs(p.id) in [12, -12, 14, -14, -16, 16]

        if p.status == 1 and not is_neutrino:
            # C√°lculo do pT
            pt = np.sqrt(p.px**2 + p.py**2)
            pt_final.append(pt)

            # C√°lculo aproximado de Eta e Phi (usando numpy)
            # Nota: Pylhe puro √†s vezes n√£o tem .eta direto, calculamos no bra√ßo:
            p_total = np.sqrt(p.px**2 + p.py**2 + p.pz**2)
            if p_total - p.pz != 0: # Evitar divis√£o por zero
                 eta = 0.5 * np.log((p_total + p.pz) / (p_total - p.pz))
            else:
                 eta = 0 # ou trate como exce√ß√£o
            eta_final.append(eta)

            phi = np.arctan2(p.py, p.px)
            phi_final.append(phi)
  return pt_final, eta_final, phi_final

# --- Processamento dos Dados ---
print("Extraindo dados cinem√°ticos...")
pt_sinal, eta_sinal, phi_sinal = var_cinematicas(eventos_sinal)
pt_fundo, eta_fundo, phi_fundo = var_cinematicas(eventos_fundo)

# --- Plotagem dos 3 Gr√°ficos (Subplots) ---
# Criamos uma figura com 1 linha e 3 colunas
fig, ax = plt.subplots(1, 3, figsize=(18, 5))

# Gr√°fico 1: Momento Transversal (pT)
ax[0].hist(pt_fundo, bins=100, range=(0, 100), alpha=0.6, color='red', label='Fundo (DY)', density=True)
ax[0].hist(pt_sinal, bins=100, range=(0, 100), alpha=0.6, color='blue', label='Sinal (WW)', density=True)
ax[0].set_xlabel(r'$p_T$ [GeV]')
ax[0].set_ylabel('Eventos Normalizados (u.a.)')
ax[0].set_title('Momento Transversal')
ax[0].legend()
ax[0].grid(alpha=0.3)

# Gr√°fico 2: Pseudorapidez (Eta)
# Eta costuma ir de -5 a 5, mas o detetor geralmente foca em -3 a 3
ax[1].hist(eta_fundo, bins=100, range=(-4, 4), alpha=0.6, color='red', label='Fundo', density=True)
ax[1].hist(eta_sinal, bins=100, range=(-4, 4), alpha=0.6, color='blue', label='Sinal', density=True)
ax[1].set_xlabel(r'$\eta$ (Pseudorapidez)')
ax[1].set_title(r'Distribui√ß√£o Angular ($\eta$)')
ax[1].legend()
ax[1].grid(alpha=0.3)

# Gr√°fico 3: √Çngulo Azimutal (Phi)
# Phi vai de -pi a +pi
ax[2].hist(phi_fundo, bins=100, range=(-np.pi, np.pi), alpha=0.6, color='red', label='Fundo', density=True)
ax[2].hist(phi_sinal, bins=100, range=(-np.pi, np.pi), alpha=0.6, color='blue', label='Sinal', density=True)
ax[2].set_xlabel(r'$\phi$ [rad]')
ax[2].set_title(r'√Çngulo Azimutal ($\phi$)')
ax[2].legend()
ax[2].grid(alpha=0.3)

plt.tight_layout() # Ajusta os espa√ßamentos para n√£o encavalar
plt.show()
```

# 4) Filtros e Cortes Cinem√°ticos

## Cortes cinem√°ticos simples: (pT> X GeV) e (|Œ∑| < Y)

```python
# --- Fun√ß√£o de An√°lise com Filtro ---
def analisar_com_corte(eventos, nome_tipo, pt_minimo=0):
    pt_lista = []

    # Contadores para responder ao Objetivo 3 e 2a
    total_eventos = len(eventos)
    eventos_que_passaram = 0
    particulas_plotadas = 0

    for e in eventos:
        passou_corte = False
        for p in e.particles:
            # Filtro b√°sico: Status 1 e n√£o ser neutrino
            is_neutrino = abs(p.id) in [12, 14, 16]

            if p.status == 1 and not is_neutrino:
                pt = np.sqrt(p.px**2 + p.py**2)

                # --- APLICA√á√ÉO DO CORTE (Objetivo 4) ---
                if pt > pt_minimo:
                    pt_lista.append(pt)
                    particulas_plotadas += 1
                    passou_corte = True

        if passou_corte:
            eventos_que_passaram += 1

    return pt_lista, total_eventos, eventos_que_passaram, particulas_plotadas

# --- Configura√ß√£o do Corte ---
corte_pt = 28 # Vamos tentar cortar tudo abaixo de 25 GeV (exemplo)

print(f"--- APLICANDO CORTE: pT > {corte_pt} GeV ---")

# Processa Sinal e Fundo
pt_sinal, ev_tot_s, ev_pass_s, part_s = analisar_com_corte(eventos_sinal, "Sinal", corte_pt)
pt_fundo, ev_tot_f, ev_pass_f, part_f = analisar_com_corte(eventos_fundo, "Fundo", corte_pt)

# --- Respostas para o Relat√≥rio (Objetivo 3 - Pergunta) ---
print(f"\nRESPOSTA SOBRE REPRESENTATIVIDADE DOS HISTOGRAMAS:")
print(f"SINAL: {ev_tot_s} eventos processados -> {part_s} part√≠culas plotadas.")
print(f"FUNDO: {ev_tot_f} eventos processados -> {part_f} part√≠culas plotadas.")
print(f"Conclus√£o: O histograma tem mais entradas que eventos pois cada evento produz m√∫ltiplos m√∫ons.\n")

print(f"--- EFICI√äNCIA DO CORTE ---")
print(f"Sinal sobrevivente: {ev_pass_s}/{ev_tot_s} ({ev_pass_s/ev_tot_s:.1%})")
print(f"Fundo sobrevivente: {ev_pass_f}/{ev_tot_f} ({ev_pass_f/ev_tot_f:.1%})")

# --- Plotagem Comparativa ---
plt.figure(figsize=(10, 6))
plt.hist(pt_fundo, bins=100, range=(0, 150), alpha=0.5, color='red', label=f'Fundo (pT > {corte_pt})', density=True)
plt.hist(pt_sinal, bins=100, range=(0, 150), alpha=0.5, color='blue', label=f'Sinal (pT > {corte_pt})', density=True)
plt.xlabel('pT [GeV]')
plt.ylabel('Densidade')
plt.title(f'Distribui√ß√£o de pT com corte de {corte_pt} GeV')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

## Novos histogramas com corte

```python
import matplotlib.pyplot as plt
import numpy as np

# --- Fun√ß√£o de Filtro (Reutilizando a l√≥gica) ---
def obter_pt_com_cortes(eventos, pt_min=20, eta_max=2.4):
    pt_antes = []
    pt_depois = []

    for e in eventos:
        passou_evento = False
        temp_pt = []

        for p in e.particles:
            # Filtro b√°sico: Status 1 e n√£o ser neutrino
            is_neutrino = abs(p.id) in [12, 14, 16]

            if p.status == 1 and not is_neutrino:
                pt = np.sqrt(p.px**2 + p.py**2)

                # Calcula Eta Manualmente
                p_total = np.sqrt(p.px**2 + p.py**2 + p.pz**2)
                if p_total - p.pz != 0:
                     eta = 0.5 * np.log((p_total + p.pz) / (p_total - p.pz))
                else:
                     eta = 999

                # Guarda TODOS para o "Antes"
                pt_antes.append(pt)

                # Verifica se passa no corte
                if (pt > pt_min) and (abs(eta) < eta_max):
                    temp_pt.append(pt)
                    passou_evento = True

        # Se o evento passou, guardamos os pTs filtrados para o "Depois"
        if passou_evento:
            pt_depois.extend(temp_pt)

    return pt_antes, pt_depois

# ==========================================
# ‚úÇÔ∏è DEFINI√á√ÉO DOS CORTES
# ==========================================
MEU_PT_MIN = 25.0  # Tenta aumentar isso se quiseres limpar mais fundo!
MEU_ETA_MAX = 2.4

print(f"Calculando histogramas para pT > {MEU_PT_MIN} e |eta| < {MEU_ETA_MAX}...")

# Processa os dados
pt_sinal_antes, pt_sinal_depois = obter_pt_com_cortes(eventos_sinal, MEU_PT_MIN, MEU_ETA_MAX)
pt_fundo_antes, pt_fundo_depois = obter_pt_com_cortes(eventos_fundo, MEU_PT_MIN, MEU_ETA_MAX)

# ==========================================
# üìä PLOTAGEM COMPARATIVA
# ==========================================
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# --- GR√ÅFICO 1: O FUNDO (Queremos ver isso sumir!) ---
ax[0].hist(pt_fundo_antes, bins=50, range=(0, 150), color='red', alpha=0.3, label='Fundo Original (Sem cortes)')
ax[0].hist(pt_fundo_depois, bins=50, range=(0, 150), color='darkred', histtype='step', linewidth=2, label='Fundo Final (Com cortes)')
ax[0].set_title(f"Impacto no BACKGROUND (Drell-Yan)\nCortes: pT>{MEU_PT_MIN}, |Œ∑|<{MEU_ETA_MAX}")
ax[0].set_xlabel("Momento Transversal ($p_T$) [GeV]")
ax[0].set_ylabel("N√∫mero de Part√≠culas")
ax[0].legend()
ax[0].grid(alpha=0.3)

# --- GR√ÅFICO 2: O SINAL (Queremos manter isso!) ---
ax[1].hist(pt_sinal_antes, bins=50, range=(0, 150), color='blue', alpha=0.3, label='Sinal Original (Sem cortes)')
ax[1].hist(pt_sinal_depois, bins=50, range=(0, 150), color='darkblue', histtype='step', linewidth=2, label='Sinal Final (Com cortes)')
ax[1].set_title(f"Impacto no SINAL (Higgs -> WW)\nCortes: pT>{MEU_PT_MIN}, |Œ∑|<{MEU_ETA_MAX}")
ax[1].set_xlabel("Momento Transversal ($p_T$) [GeV]")
ax[1].legend()
ax[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# --- AN√ÅLISE DE EFICI√äNCIA ---
print("\n=== RELAT√ìRIO DE SOBREVIV√äNCIA ===")
print(f"üî¥ Fundo Restante: {len(pt_fundo_depois)} de {len(pt_fundo_antes)} part√≠culas ({len(pt_fundo_depois)/len(pt_fundo_antes):.1%})")
print(f"üîµ Sinal Restante: {len(pt_sinal_depois)} de {len(pt_sinal_antes)} part√≠culas ({len(pt_sinal_depois)/len(pt_sinal_antes):.1%})")

if (len(pt_fundo_depois)/len(pt_fundo_antes)) < (len(pt_sinal_depois)/len(pt_sinal_antes)):
    print("\n‚úÖ CONCLUS√ÉO: O corte foi eficaz! Removeu mais fundo do que sinal.")
else:
    print("\n‚ö†Ô∏è CONCLUS√ÉO: O corte n√£o foi muito seletivo. Tente aumentar o pT m√≠nimo.")
```

## Vari√°vel de Mandelstam "s"

```python
print("="*80)
print("VERIFICA√á√ÉO DA VARI√ÅVEL DE MANDELSTAM 'S'")
print("="*80)

def calcular_mandelstam_s(evento, indice):
    E_tot, px_tot, py_tot, pz_tot = 0.0, 0.0, 0.0, 0.0

    particulas_usadas = []

    for p in evento.particles:
        # Apenas estado final (1) e ignorar neutrinos (12, 14, 16)
        is_neutrino = abs(p.id) in [12, 14, 16]
        if p.status == 1 and not is_neutrino:
            E_tot += p.e
            px_tot += p.px
            py_tot += p.py
            pz_tot += p.pz
            particulas_usadas.append(f"{p.id}")

    # s = E^2 - p^2
    p_sq = px_tot**2 + py_tot**2 + pz_tot**2
    s = E_tot**2 - p_sq

    # sqrt(s) √© a energia de centro de massa reconstru√≠da
    sqrt_s = np.sqrt(s) if s > 0 else 0

    print(f"Evento {indice}: sqrt(s) = {sqrt_s:.2f} GeV | Part√≠culas usadas: {particulas_usadas}")

# Indices solicitados: 1 + 1000*i para i de 0 a 9
indices = [1 + 1000 * i for i in range(10)]

print("\n--- Amostra de SINAL ---")
for idx in indices:
    if idx < len(eventos_sinal):
        calcular_mandelstam_s(eventos_sinal[idx], idx)

print("\n--- Amostra de FUNDO ---")
for idx in indices:
    if idx < len(eventos_fundo):
        calcular_mandelstam_s(eventos_fundo[idx], idx)
```

## HISTOGRAMAS DE MASSA INVARIANTE DO PAR (DIMUON)

```python
import itertools

def calcular_massa_pares(eventos, label_tipo):
    massas_pares = []

    for e in eventos:
        muons = []
        # Selecionar apenas M√∫ons (ID 13 ou -13) do estado final (status 1)
        for p in e.particles:
            if p.status == 1 and abs(p.id) == 13:
                muons.append(p)

        # Se houver pelo menos 2 m√∫ons, formamos o par
        if len(muons) >= 2:
            # Ordenar por pT para pegar os dois mais energ√©ticos (leading muons) se houver mais de 2
            muons.sort(key=lambda p: (p.px**2 + p.py**2), reverse=True)
            m1 = muons[0]
            m2 = muons[1]

            # Soma dos quadrivetores (E, px, py, pz)
            E_par = m1.e + m2.e
            px_par = m1.px + m2.px
            py_par = m1.py + m2.py
            pz_par = m1.pz + m2.pz

            # Massa Invariante: M^2 = E^2 - |p|^2
            m_sq = E_par**2 - (px_par**2 + py_par**2 + pz_par**2)

            if m_sq > 0:
                massas_pares.append(np.sqrt(m_sq))

    return massas_pares

print("Calculando massa invariante de pares de m√∫ons...")
massas_par_sinal = calcular_massa_pares(eventos_sinal, "Sinal")
massas_par_fundo = calcular_massa_pares(eventos_fundo, "Fundo")

# --- Plotagem dos Histogramas de Pares ---
plt.figure(figsize=(10, 6))

# Plot do Fundo (Drell-Yan -> 2 mu)
plt.hist(massas_par_fundo, bins=60, range=(0, 150),
         color='red', alpha=0.4, density=True, label=r'Fundo (Par $\mu^+\mu^-$)')

# Plot do Sinal (Higgs -> WW -> 2 mu 2 nu)
plt.hist(massas_par_sinal, bins=60, range=(0, 150),
         color='blue', histtype='step', linewidth=2, density=True, label=r'Sinal (Par $\mu^+\mu^-$)')

plt.title('Distribui√ß√£o de Massa Invariante do Par de M√∫ons (Normalizado p/ √Årea)')
plt.xlabel(r'Massa Invariante $m_{\mu\mu}$ [GeV]')
plt.ylabel('Densidade de Eventos')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

print(f"Eventos com pares de m√∫ons encontrados: SINAL={len(massas_par_sinal)}, FUNDO={len(massas_par_fundo)}")
```

## Histogramas 2D (mapa de calor) de massa invariante versus Œ∑

```python
def reconstruir_massa_eta(eventos):
    massas = []
    etas_sistema = [] # Eta do sistema combinado

    for e in eventos:
        E_tot, px_tot, py_tot, pz_tot = 0.0, 0.0, 0.0, 0.0
        n_particulas = 0

        for p in e.particles:
            # Filtro: status 1 e ignorar neutrinos
            is_neutrino = abs(p.id) in [12, 14, 16]
            if p.status == 1 and not is_neutrino:
                E_tot += p.e
                px_tot += p.px
                py_tot += p.py
                pz_tot += p.pz
                n_particulas += 1

        # S√≥ calcula se houver pelo menos 2 part√≠culas detect√°veis
        if n_particulas >= 2:
            p_sq = px_tot**2 + py_tot**2 + pz_tot**2
            m_sq = E_tot**2 - p_sq
            # Evita raiz de n√∫mero negativo devido a precis√£o num√©rica
            massa = np.sqrt(max(0.0, m_sq))
            massas.append(massa)

            # Calcular pseudorapidez (eta) do sistema combinado
            # F√≥rmula: eta = 0.5 * ln( (|p| + pz) / (|p| - pz) )
            p_mod = np.sqrt(p_sq)
            denom = p_mod - pz_tot

            # Prote√ß√£o contra divis√£o por zero
            # Se o denominador for muito pequeno (menor que 1e-10), definimos um eta 'grande'
            if abs(denom) > 1e-10:
                # Prote√ß√£o extra: argumento do log deve ser positivo
                arg = (p_mod + pz_tot) / denom
                if arg > 0:
                    eta = 0.5 * np.log(arg)
                else:
                    eta = 0 # Fallback caso f√≠sico improv√°vel para sistema reconstru√≠do
            else:
                # Se pz ~ |p|, a part√≠cula est√° muito pr√≥xima do eixo z -> eta tende a infinito
                # O sinal depende de pz ser positivo ou negativo
                eta = 10e10 if pz_tot > 0 else -10e10

            etas_sistema.append(eta)

    return massas, etas_sistema

print("Reconstruindo Massa Invariante e Eta com prote√ß√£o num√©rica...")
massa_sinal, eta_sis_sinal = reconstruir_massa_eta(eventos_sinal)
massa_fundo, eta_sis_fundo = reconstruir_massa_eta(eventos_fundo)

# --- Histogramas 2D (Mapa de Calor) ---
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# Sinal
# Usamos range limitado em eta para focar na regi√£o central do detector
h1 = ax[0].hist2d(massa_sinal, eta_sis_sinal, bins=50, range=[[0, 200], [-6, 6]], cmap='plasma')
ax[0].set_title(r'SINAL (Higgs): Massa Inv. vs $\eta$')
ax[0].set_xlabel(r'Massa Invariante (GeV)')
ax[0].set_ylabel(r'$\eta$ do sistema')
fig.colorbar(h1[3], ax=ax[0])

# Fundo
h2 = ax[1].hist2d(massa_fundo, eta_sis_fundo, bins=50, range=[[0, 200], [-6, 6]], cmap='plasma')
ax[1].set_title(r'FUNDO (DY): Massa Inv. vs $\eta$')
ax[1].set_xlabel(r'Massa Invariante (GeV)')
ax[1].set_ylabel(r'$\eta$ do sistema')
fig.colorbar(h2[3], ax=ax[1])

plt.show()
```

# [7] GR√ÅFICOS E VISUALIZA√á√ÉO COM NORMALIZA√á√ÉO

## Histogramas normalizados pelo Yield

```python
# 1. PAR√ÇMETROS DE NORMALIZA√á√ÉO
# ------------------------------------------------------------------------------
sigma_sinal = 0.4395    # pb (Higgs -> WW)
sigma_fundo = 841.64    # pb (Drell-Yan -> mumu)
luminosidade = 10000.0  # 10 fb^-1

# Pesos (Weights)
n_sinal = len(eventos_sinal)
n_fundo = len(eventos_fundo)

w_sinal = (sigma_sinal * luminosidade) / n_sinal
w_fundo = (sigma_fundo * luminosidade) / n_fundo

print(r"--- PAR√ÇMETROS ---")
print(f"Peso Sinal: {w_sinal:.5f} | Peso Fundo: {w_fundo:.5f}")

# 2. FUN√á√ïES DE EXTRA√á√ÉO
# ------------------------------------------------------------------------------
def get_kinematics(eventos):
    """Retorna listas de pT, eta, phi para todos os m√∫ons do evento."""
    pts, etas, phis = [], [], []
    for e in eventos:
        for p in e.particles:
            if p.status == 1 and abs(p.id) == 13: # M√∫ons
                pts.append(np.sqrt(p.px**2 + p.py**2))
                phis.append(np.arctan2(p.py, p.px))
                pm = np.sqrt(p.px**2 + p.py**2 + p.pz**2)
                if abs(pm - p.pz) > 1e-9:
                    etas.append(0.5 * np.log((pm + p.pz)/(pm - p.pz)))
                else:
                    etas.append(0)
    return pts, etas, phis

def get_mass_pair(eventos):
    """Retorna lista de massas invariantes do par de m√∫ons (1 por evento)."""
    massas = []
    for e in eventos:
        muons = [p for p in e.particles if p.status == 1 and abs(p.id) == 13]
        if len(muons) >= 2:
            muons.sort(key=lambda p: p.px**2 + p.py**2, reverse=True)
            m1, m2 = muons[0], muons[1]
            E = m1.e + m2.e
            px = m1.px + m2.px
            py = m1.py + m2.py
            pz = m1.pz + m2.pz
            m2_sys = E**2 - (px**2 + py**2 + pz**2)
            if m2_sys > 0: massas.append(np.sqrt(m2_sys))
    return massas

# Executando extra√ß√£o
print("Processando dados...")
pt_s, eta_s, phi_s = get_kinematics(eventos_sinal)
pt_f, eta_f, phi_f = get_kinematics(eventos_fundo)

m_sinal = get_mass_pair(eventos_sinal)
m_fundo = get_mass_pair(eventos_fundo)

# Pesos (devem ter o tamanho dos dados extra√≠dos)
w_pt_s, w_pt_f = [w_sinal]*len(pt_s), [w_fundo]*len(pt_f)
w_eta_s, w_eta_f = [w_sinal]*len(eta_s), [w_fundo]*len(eta_f)
w_phi_s, w_phi_f = [w_sinal]*len(phi_s), [w_fundo]*len(phi_f)
w_m_s, w_m_f = [w_sinal]*len(m_sinal), [w_fundo]*len(m_fundo)

# 3. PLOTAGEM 2x2 (pT, Eta, Phi, Massa)
# ------------------------------------------------------------------------------
plt.rcParams.update({'font.size': 12})
fig, ax = plt.subplots(2, 2, figsize=(16, 12), dpi=100)

# A) Momento Transversal (pT)
ax[0,0].hist(pt_f, bins=50, range=(0, 150), weights=w_pt_f, color='red', alpha=0.5, label='Fundo')
ax[0,0].hist(pt_s, bins=50, range=(0, 150), weights=w_pt_s, color='blue', alpha=0.5, label='Sinal')
ax[0,0].set_title(r'Momento Transversal ($p_T$)')
ax[0,0].set_xlabel(r'$p_T$ [GeV]')
ax[0,0].set_ylabel(r'Eventos / ano')
ax[0,0].set_yscale('log') # Log mantido para ver o sinal
ax[0,0].legend()
ax[0,0].grid(True, alpha=0.3)

# B) Pseudorapidez (Eta)
ax[0,1].hist(eta_f, bins=40, range=(-5, 5), weights=w_eta_f, color='red', alpha=0.5, label='Fundo')
ax[0,1].hist(eta_s, bins=40, range=(-5, 5), weights=w_eta_s, color='blue', alpha=0.5, label='Sinal')
ax[0,1].set_title(r'Pseudorapidez ($\eta$)')
ax[0,1].set_xlabel(r'$\eta$')
ax[0,1].set_yscale('log')
ax[0,1].grid(True, alpha=0.3)

# C) √Çngulo Azimutal (Phi)
ax[1,0].hist(phi_f, bins=30, range=(-3.2, 3.2), weights=w_phi_f, color='red', alpha=0.5, label='Fundo')
ax[1,0].hist(phi_s, bins=30, range=(-3.2, 3.2), weights=w_phi_s, color='blue', alpha=0.5, label='Sinal')
ax[1,0].set_title(r'√Çngulo Azimutal ($\phi$)')
ax[1,0].set_xlabel(r'$\phi$ [rad]')
ax[1,0].set_yscale('log')
ax[1,0].grid(True, alpha=0.3)

# D) Massa Invariante (LINEAR, N√ÉO-STACKED)
# Aqui atendemos ao pedido: Massa na parte 1, sem ser log (ou linear para compara√ß√£o)
ax[1,1].hist(m_fundo, bins=60, range=(60, 160), weights=w_m_f,
             color='red', alpha=0.5, label='Fundo (DY)')
ax[1,1].hist(m_sinal, bins=60, range=(60, 160), weights=w_m_s,
             color='blue', alpha=0.5, label='Sinal (Higgs)')

ax[1,1].set_title(r'Massa Invariante (Escala Linear)')
ax[1,1].set_xlabel(r'$m_{\mu\mu}$ [GeV]')
ax[1,1].set_ylabel(r'Eventos / ano')
# ax[1,1].set_yscale('linear') # Padr√£o √© linear
ax[1,1].legend()
ax[1,1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(r"Nota: No gr√°fico de Massa (linear), o Sinal (azul) √© praticamente invis√≠vel")
print(r"devido √† enorme diferen√ßa de escala em rela√ß√£o ao Fundo (vermelho).")
```

## Distribui√ß√£o de massa invariante empilhada (stacked).

```python
# 1. RECONSTRU√á√ÉO DA MASSA DO SISTEMA
# ------------------------------------------------------------------------------
def get_mass_pair(eventos):
    massas = []
    for e in eventos:
        # Filtra apenas m√∫ons (13)
        muons = [p for p in e.particles if p.status == 1 and abs(p.id) == 13]
        if len(muons) >= 2:
            # Ordena por pT (leading muons)
            muons.sort(key=lambda p: p.px**2 + p.py**2, reverse=True)
            m1, m2 = muons[0], muons[1]
            E = m1.e + m2.e
            px = m1.px + m2.px
            py = m1.py + m2.py
            pz = m1.pz + m2.pz
            m2_sys = E**2 - (px**2 + py**2 + pz**2)
            if m2_sys > 0:
                massas.append(np.sqrt(m2_sys))
    return massas

m_sinal = get_mass_pair(eventos_sinal)
m_fundo = get_mass_pair(eventos_fundo)

# Pesos para massa
w_m_s = [w_sinal] * len(m_sinal)
w_m_f = [w_fundo] * len(m_fundo)

# 2. PLOTAGEM EMPILHADA (STACKED)
# ------------------------------------------------------------------------------
plt.figure(figsize=(10, 8), dpi=100)

# Dados e pesos como listas de listas para o stack
# Ordem: [Fundo, Sinal] -> Fundo na base, Sinal no topo
dados = [m_fundo, m_sinal]
pesos = [w_m_f, w_m_s]
cores = ['red', 'blue']
labels = [r'Fundo ($Z/\gamma^* \rightarrow \mu\mu$)', r'Sinal ($H \rightarrow WW$)']

plt.hist(dados,
         bins=60,
         range=(60, 160),
         weights=pesos,
         stacked=True,          # Empilha os histogramas
         color=cores,
         label=labels,
         edgecolor='black',
         linewidth=0.5,
         alpha=0.8)

# Decora√ß√£o e Escalas
plt.title(rf'Massa Invariante de Dim√∫ons (Normalizado para $\mathcal{{L}} = 10\ \mathrm{{fb}}^{{-1}}$)', fontsize=16)
plt.xlabel(r'Massa Invariante $m_{\mu\mu}$ [GeV]', fontsize=14)
plt.ylabel(r'Eventos / bin', fontsize=14)
plt.yscale('log') # Escala Log para visualizar a diferen√ßa de ordens de grandeza
plt.legend(fontsize=12)
plt.grid(True, which='both', linestyle='--', alpha=0.5)

# Marcas de refer√™ncia f√≠sica
plt.text(91, 1e5, r'$\mathbf{Z^0}$', color='maroon', ha='center', fontsize=14, fontweight='bold')
plt.axvline(125, color='green', linestyle='--', linewidth=2, label='Higgs (125 GeV)')
plt.text(126, 20, r'$m_H$', color='green', fontweight='bold', fontsize=14)

plt.tight_layout()
plt.show()

# Coment√°rio final sobre a estrutura
print(r"")
print(r"Nota: O gr√°fico utiliza escala logar√≠tmica (eixo Y) porque o Fundo √© ~2000x maior que o Sinal.")
print(r"O pico em ~91 GeV √© o B√≥som Z (Fundo). O sinal do Higgs aparece como um excesso largo √† esquerda de 125 GeV.")
```
